```python
import matplotlib.pyplot as plt
import random
import numpy as np

def simulate_chapman_cycle(O2, O, O3, k1, k2, k3, k4, max_steps=100000, equilibrium_threshold=0.01):
    O2_counts = [O2]
    O_counts = [O]
    O3_counts = [O3]
    time_steps = [0]
    for step in range(1, max_steps + 1):
        rates = {
            ('O2_photolysis'): k1 * O2,          #O  2O
            ('ozone_formation'): k2 * O * O2,    #O + O  O
            ('ozone_photolysis'): k3 * O3,       #O  O + O
            ('ozone_depletion'): k4 * O * O3     #O + O  2O
        }
        
        possible_rates = {k: v for k, v in rates.items() if v > 0}
        if not possible_rates:
            break
            
        total_rate = sum(possible_rates.values())
        reaction = random.choices(list(possible_rates.keys()), weights=list(possible_rates.values()), k=1)[0]
        
        if reaction == 'O2_photolysis':
            O2 -= 1
            O += 2
        elif reaction == 'ozone_formation':
            O -= 1
            O2 -= 1
            O3 += 1
        elif reaction == 'ozone_photolysis':
            O3 -= 1
            O += 1
            O2 += 1
        elif reaction == 'ozone_depletion':
            O -= 1
            O3 -= 1
            O2 += 2
        
        O2 = max(O2, 0)
        O = max(O, 0)
        O3 = max(O3, 0)
        O2_counts.append(O2)
        O_counts.append(O)
        O3_counts.append(O3)
        time_steps.append(step)

        if step > 10:
            recent_changes = [
                max(O2_counts[-10:]) - min(O2_counts[-10:]),
                max(O_counts[-10:]) - min(O_counts[-10:]),
                max(O3_counts[-10:]) - min(O3_counts[-10:])
            ]
            if all(change < equilibrium_threshold for change in recent_changes):
                break
                
    return time_steps, O2_counts, O_counts, O3_counts

def plot_chapman_results(time_steps, O2, O, O3):
    plt.figure(figsize=(10, 6))
    plt.plot(time_steps, O2, label='O', linewidth=2)
    plt.plot(time_steps, O, label='O atoms', linewidth=2)
    plt.plot(time_steps, O3, label='O (ozone)', linewidth=2)
    plt.xlabel('Time (simulation steps)')
    plt.ylabel('Relative Concentration')
    plt.title('Chapman Cycle Simulation')
    plt.legend()
    plt.grid(True)
    plt.show()

initial_O2 = 1000    #Abundant in the atmosphere
initial_O = 1        #Rare in the atmosphere
initial_O3 = 10      #Somewhat rare in the atmosphere
k1 = 0.00002         #O photolysis (very slow)
k2 = 0.05            #O formation (fast)
k3 = 0.005           #O photolysis (medium)
k4 = 0.001           #O depletion (slow)

time_steps, O2, O, O3 = simulate_chapman_cycle(initial_O2, initial_O, initial_O3, k1, k2, k3, k4,max_steps=100000)
plot_chapman_results(time_steps, O2, O, O3)

final_O3_O2_ratio = O3[-1] / O2[-1]
print("Equilibrium:")
print(f"O/O ratio: {final_O3_O2_ratio:.4f}")
print(f"O atoms concentration: {O[-1]:.2f}")
```

```python
import random
from itertools import combinations_with_replacement
import matplotlib.pyplot as plt

def simulate_particles(A, B, C, max_steps=1000000):
    for _ in range(max_steps):
        possible_pairs = []
        weights = []
        if A > 0 and B > 0:
            possible_pairs.append(('A', 'B'))
            weights.append(A * B)
        if A > 0 and C > 0:
            possible_pairs.append(('A', 'C'))
            weights.append(A * C)
        if B > 0 and C > 0:
            possible_pairs.append(('B', 'C'))
            weights.append(B * C)
        if not possible_pairs:
            return True
        total_weight = sum(weights)
        probabilities = [w/total_weight for w in weights]
        collision = random.choices(possible_pairs, weights=probabilities, k=1)[0]

        if collision == ('A', 'B'):
            A -= 1
            B -= 1
            C += 2
        elif collision == ('A', 'C'):
            A -= 1
            C -= 1
            B += 2
        elif collision == ('B', 'C'):
            B -= 1
            C -= 1
            A += 2
    return False

def check_congruence(a, b, c):
    return (a % 3 == b % 3) or (a % 3 == c % 3) or (b % 3 == c % 3)

def analyze_system(max_value=5):
    results = {
        'congruent': {'terminates': 0, 'not_terminates': 0},
        'non_congruent': {'terminates': 0, 'not_terminates': 0}
    }
    detailed_results = []
    
    for a, b, c in combinations_with_replacement(range(max_value + 1), 3):
        if a + b + c == 0:
            continue
        congruent = check_congruence(a, b, c)
        terminates = 0
        if simulate_particles(a, b, c):
            terminates += 1
        if terminates > 0:
            status = 'Terminated'
        else: 
            status = 'Did not terminate'
        detailed_results.append((a, b, c, congruent, status))
        if congruent:
            key = 'congruent'
        else:
            key = 'non_congruent'
        if status == 'Terminated':
            subkey = 'terminates'
        else: 
            subkey = 'not_terminates'
        results[key][subkey] += 1
    
    return results, detailed_results

def plot_results(results):
    labels = ['Congruent mod 3', 'Non-congruent']
    terminate_counts = [results['congruent']['terminates'], results['non_congruent']['terminates']]
    non_terminate_counts = [results['congruent']['not_terminates'], results['non_congruent']['not_terminates']]
    x = range(len(labels))
    fig, ax = plt.subplots()
    rects1 = ax.bar(x, terminate_counts, label= 'Terminates')
    rects2 = ax.bar(x, non_terminate_counts, bottom=terminate_counts, label='Does Not Terminate')
    ax.set_ylabel('Number of Unique Configurations')
    ax.set_title('Termination by Congruence Modulo 3')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    fig.tight_layout()
    plt.show()

def print_detailed_results(detailed_results):
    print("\nDetailed Results for Each Configuration (A  B  C):")
    print("Format: (A, B, C) | Congruent? | Status")
    print("-" * 50)
    for result in sorted(detailed_results, key=lambda x: (x[0], x[1], x[2])):
        a, b, c, congruent, status = result
        if congruent:
            cong_status = "Yes"
        else:
            cong_status = "No"
        print(f"({a:2d}, {b:2d}, {c:2d}) | {cong_status:8s} | {status}")

max_value = 6
results, detailed_results = analyze_system(max_value)

plot_results(results)

print(f"\nTesting all unique combinations where A  B  C  {max_value}")
print("\nDetailed Breakdown Statistics:")
print("\nCongruent mod 3 configurations:")
print(f"Terminates: {results['congruent']['terminates']}")
print(f"Does not terminate: {results['congruent']['not_terminates']}")

print("\nNon-congruent configurations:")
print(f"Terminates: {results['non_congruent']['terminates']}")
print(f"Does not terminate: {results['non_congruent']['not_terminates']}")

print_detailed_results(detailed_results)
```
